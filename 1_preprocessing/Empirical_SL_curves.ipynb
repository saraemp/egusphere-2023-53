# This code: 
Extracts data from Miller et al. (2020) that correspond to the ice-free Paleocene, ca. 66-56 Ma (greenhouse) and the ice-dominated Oligocene ca. 33.9-23.9 Ma (icehouse) periods

```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
```

```python
Miller_int=pd.read_csv('./Miller_et_al_2020_interp1d_age_sea_level.csv')
```

```python
plt.plot(Miller_int.age_ka, Miller_int.sea_level_m)
plt.ylabel('Elevation (m)')
plt.xlabel('Time, (ka)')
```

```python
GH=np.append(np.zeros(20),Miller_int.sea_level_m[(Miller_int.age_ka>=56e3)&(Miller_int.age_ka<=66e3)])
c2=np.append(np.arange(0,2e6,1e5), np.arange(2e6,12.05e6,50e3))
plt.plot(GH,c2, color='green')

IH=np.append(np.zeros(20),Miller_int.sea_level_m[(Miller_int.age_ka>=23.9e3)&(Miller_int.age_ka<=33.9e3)])
c2=np.append(np.arange(0,2e6,1e5), np.arange(2e6,12.05e6,50e3))
plt.plot(IH, c2, color='blue')

IH_df = pd.DataFrame(IH, c2)
IH_df.to_csv('../2_model_running/data/SL_IH.csv', float_format='%.3f', header=False, sep=' ')

GH_df = pd.DataFrame(GH, c2)
GH_df.to_csv('../2_model_running/data/SL_GH.csv', float_format='%.3f', header=False, sep=' ')
```
